```python
"""
Establezca 2 modelos de clasificación para el data Iirs
"""
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
iris_d = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/iris.csv')
display(iris_d.head())
```

```python
#LabelEncoder
Encoder1 = LabelEncoder()
Name_iris = Encoder1.fit_transform(iris_d['Name'])
#Tabla
iris_d = iris_d.drop(columns = 'Name')
iris_d['Name'] = Name_iris
display(iris_d.head())
```

```python
x_d = iris_d.drop(columns = 'Name')
y_d = iris_d['Name']
```

```python
display(x_d.head())
```

```python
from sklearn.linear_model import LinearRegression
def calcular_vif (df, características):     
    vif, tolerancia = {}, {}
    
    for característica in características: 
        
        x = [f for f in características if f != característica]         
        x, y = df[x], df[característica]
         
        r2 = LinearRegression().fit(x, y).score(x, y)                 
        
        
        tolerancia[característica] = 1 - r2
        
        vif[característica] = 1/(tolerancia[característica])
    
    return pd.DataFrame({'VIF': vif, 'Tolerancia': tolerancia})
```

```python
calcular_vif(df = x_d,características = ('SepalLength',
                                          'SepalWidth'	,
                                            #'PetalLength',
                                             'PetalWidth'
                                          ))
```

```python
x_d = iris_d.drop(columns = ['Name','PetalLength'])
x = np.array(x_d)
y = np.array(y_d)
```

```python
import matplotlib.pyplot as plt
pd.plotting.scatter_matrix(x_d,figsize = (15,10) , diagonal = 'kde')
plt.show()
```

```python
x_d = iris_d.drop(columns = ['Name','PetalLength','PetalWidth'])
x = np.array(x_d)
```

```python
from sklearn.preprocessing import MinMaxScaler as MMS
scaler = MMS(feature_range = (0,1))
x = scaler.fit_transform(x)
```

```python
print(np.shape(x))
y = np.reshape(y,(len(y),1))
print(np.shape(y))
```

```python
from sklearn.model_selection import train_test_split as tts
xtrain,xtest,ytrain,ytest = tts(x,y,test_size = 0.2, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.ensemble import GradientBoostingClassifier as GBC
from sklearn.metrics import recall_score
modelRFC = RFC(random_state = 42)
modelGBC = GBC(random_state = 42)

recall_score_train = []
recall_score_test = []

for i in [modelRFC,modelGBC]:
  i.fit(xtrain,ytrain)
  ypredtrain = i.predict(xtrain)
  ypredtest = i.predict(xtest)

  recallscoretrain = recall_score(ytrain,ypredtrain,average = 'micro')
  recallscoretest = recall_score(ytest,ypredtest,average = 'micro')

  recall_score_train.append(recallscoretrain)
  recall_score_test.append(recallscoretest)

dicc = {'Modelo':['Random Forest Classifier','Gradient Boosting Classifier']}
dicc_d = pd.DataFrame(dicc)
dicc_d['recall score train'] = recall_score_train
dicc_d['recall score test'] = recall_score_test
display(dicc_d)
```

```python
"""
¿Existe sobreajuste al aplicar un modelo de RF con n = 200 para el modelo de wine.csv? 
"""
from sklearn.metrics import r2_score
wine_d = pd.read_excel('/content/wine.xlsx')
display(wine_d)
```

```python
W_d = wine_d.drop(columns = ['Unnamed: 0','description','designation','region_1','region_2','winery'])
W_d = W_d.dropna()
display(W_d)
```

```python
encoder = LabelEncoder()
X_d = W_d.drop(columns = ['price'])

labelencoder = LabelEncoder()
country = labelencoder.fit_transform(np.array(W_d['country']))
points = labelencoder.fit_transform(np.array(W_d['points']))
province = labelencoder.fit_transform(np.array(W_d['province']))
variety = labelencoder.fit_transform(np.array(W_d['variety']))


X_d = X_d.drop(columns = ['points','province','variety'])
X_d['country'] = country
X_d['points'] = points
X_d['province'] = province
X_d['variety'] = variety



X_array = np.array(X_d)

y_d = W_d['price']
y_array = np.array(y_d)
```

```python
print(np.shape(X_array))
print(np.shape(y_array))
y_array = np.reshape(y_array,(len(y_array),1))
print(np.shape(y_array))
```

```python
scaler = MMS(feature_range = (0,1))
X_array = scaler.fit_transform(X_array)
```

```python
Xtrain,Xtest,ytrain,ytest = tts(X_array,y_array, random_state = 42, test_size = 0.2) #dividimos la data

from sklearn.ensemble import RandomForestRegressor as RFR
modelRFR = RFR(random_state = 42, n_estimators =200)

modelRFR.fit(Xtrain,ytrain) #Entrenamos el modelo

#verificamos los resultados
ypredtest = modelRFR.predict(Xtest)
ypredtrain = modelRFR.predict(Xtrain)

r2scoretest = r2_score(ytest,ypredtest)
r2scoretrain = r2_score(ytrain,ypredtrain)
```

```python
dicc_w = {'Modelo':['Random Forest Regressor'],'r2 score train':[r2scoretrain],'r2 score test':[r2scoretest]}
dicc_w_d = pd.DataFrame(dicc_w)
display(dicc_w_d)
print('Modelo mínimamente generalizado')
```

```python
from sklearn.model_selection import train_test_split as tts
x2_ctrain,x2_ctest,y2_ctrain,y2_ctest = tts(x2_c,y2_c,test_size = 0.2,random_state = 42)
```

```python
import tensorflow as tf
ModeloClasificador = tf.keras.Sequential([
        tf.keras.Input(shape = (13,)),
        tf.keras.layers.Dense(13,activation = 'relu'),
        tf.keras.layers.Dense(60,activation = 'relu'),
        tf.keras.layers.Dense(30,activation = 'relu'),
        tf.keras.layers.Dense(10,activation = 'relu'),
        tf.keras.layers.Dense(4,activation = 'relu'),
        tf.keras.layers.Dense(2,activation = 'relu'),
        tf.keras.layers.Dense(1,activation = 'softmax')
       
])

ModeloClasificador.compile(loss = tf.keras.losses.binary_crossentropy,
                           optimizer = tf.keras.optimizers.SGD(),
                           metrics = 'accuracy')
historico = ModeloClasificador.fit(x2_ctrain,y2_ctrain, epochs = 500,verbose = False)
```

```python
pd.DataFrame(historico.history).plot()
```

```python
ypredtrain2 = np.round(ModeloClasificador.predict(x2_ctrain),0)
ypredtest2 = np.round(ModeloClasificador.predict(x2_ctest),0)

rcstrain = recall_score(y2_ctrain,ypredtrain2)
rcstest = recall_score(y2_ctest,ypredtest2)

dicc1 = {'Modelo':['GradientBoostingClassifier'],'recall score train':[rcstrain],'recall score test':[rcstest]}
dicc1_d = pd.DataFrame(dicc1)
display(dicc1_d)
```
